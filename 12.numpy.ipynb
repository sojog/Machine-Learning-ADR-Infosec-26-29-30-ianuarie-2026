# Fast mathematical operations over arrays

```python
#  An array object of arbitrary homogeneous items

arr = np.array([1, 2, 3, 4, 5, 6])
arr 
```

```python
arr + 3
```

```python
arr ** 2
```

```python
arr + arr
```

```python

arr1 = np.array([1, 2, 3, 4, 5, 6])
arr2 = np.array([1, 2, 3, 4, 5])
arr1 + arr2 
```

Map - filter - reduce

# 1. Map 

```python
arr = np.array([1, 2, 3, 4, 5, 6])

arr - 3
```

# 2. Filter

```python
arr = np.array([1, 2, 3, 4, 5, 6])

arr > 3
```

```python
arr[arr > 3]
```

```python
arr[arr % 2 == 0]
```
