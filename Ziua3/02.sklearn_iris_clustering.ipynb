```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
iris_dataset = load_iris()
iris_dataset
```

```python
type(iris_dataset)
```

```python
print(iris_dataset.DESCR)
```

```python
iris_dataset["data"]
```

```python
sepal_lenght = iris_dataset["data"][ :, 0 ]
sepal_lenght
```

```python
sepal_width = iris_dataset["data"][ :, 1 ]
sepal_width
```

```python
plt.scatter(sepal_lenght, sepal_width)
plt.xlabel("Lungimea sepalei")
plt.ylabel("Latimea sepalei")
```

```python
model_clustering = KMeans(n_clusters=3)
model_clustering
```

```python
x = iris_dataset["data"]

model_clustering.fit(x )
```

```python
model_clustering.cluster_centers_
```

```python
model_clustering.labels_
```

```python
plt.scatter(sepal_lenght, sepal_width, c=model_clustering.labels_)
plt.title("KMeans - prezi (sepala)")
plt.xlabel("Lungimea sepalei")
plt.ylabel("Latimea sepalei")
```

```python
plt.scatter(sepal_lenght, sepal_width, c=iris_dataset["target"])
plt.title("Valori reale (sepala)");
```

```python
petal_length = iris_dataset["data"][ :, 2 ]
petal_width = iris_dataset["data"][ :, 3 ]

plt.scatter(petal_length, petal_width, c=model_clustering.labels_)
plt.title("KMeans - prezi (petala)")
plt.xlabel("Lungimea sepalei")
plt.ylabel("Latimea sepalei")
```

```python
plt.scatter(petal_length, petal_width, c=iris_dataset["target"])
plt.title("Petale - varianta reala")
plt.legend(iris_dataset["target_names"]);
```
