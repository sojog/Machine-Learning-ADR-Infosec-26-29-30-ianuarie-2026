# Descoperire Functie AND pentru Neuron

```python
def functie_activare(v):
    if v < 0:
        return 0
    return 1

def functie_neuron_and(x1, x2):
    w1 = 0.5
    w2 = 0.5
    b = -1
    rezultat = functie_activare(x1 * w1 + x2 * w2 + b)
    print(f"{x1} and {x2} = {rezultat}")
```

```python
functie_neuron_and(1, 1)
functie_neuron_and(1, 0)
functie_neuron_and(0, 1)
functie_neuron_and(0, 0)
```

```python
AND_CLASSIC(1, 1)
AND_CLASSIC(1, 0)
AND_CLASSIC(0, 1)
AND_CLASSIC(0, 0)
```

# Descoperire functie OR pentru Neuron

```python
def OR_CLASSIC(x1, x2):
    print(f"{x1} and {x2} = {x1 or x2}")
```

```python
OR_CLASSIC(1, 1)
OR_CLASSIC(1, 0)
OR_CLASSIC(0, 1)
OR_CLASSIC(0, 0)
```
