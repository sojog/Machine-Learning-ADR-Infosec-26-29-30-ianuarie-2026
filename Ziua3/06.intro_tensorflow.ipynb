```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 11)
x
```

```python
y = 2 * x + 1
x
```

```python
plt.scatter(x, y)
plt.plot(x, y)
#
```

```python
model = tf.keras.models.Sequential(
    # Layer. de Output
    [tf.keras.layers.Dense(1, input_shape=(1,))]
)
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['mae'])
```

```python
model.fit(x, y,epochs=10)
```

```python
model.predict(np.array([11]))
```

```python
model.predict(np.array([2]))
```

```python
model_2 = tf.keras.models.Sequential(
    # Layer. de Output
    [tf.keras.layers.Dense(1, input_shape=(1,))]
)
```

```python
model_2.compile(optimizer='sgd', loss='mse', metrics=['mae'])
```

```python
model_2.fit(x, y, epochs=300, verbose=0)
```

```python
model_2.predict(np.array([2]))
```

```python
model_3 = tf.keras.models.Sequential(
    # Layer. de Output
    [tf.keras.layers.Dense(1, input_shape=(1,))]
)
```

```python
model_3.compile(optimizer='sgd', loss='mse', metrics=['mae'])
```

```python
model_3.fit(x, y, epochs=1000, verbose=0)
```

```python
model_3.predict(np.array([2]))
```

```python
model_3.predict(np.array([11]))
```

# ! Spre deosebire de SKLEARN avem nevoie de compilare

```python
model_4_secvential = tf.keras.models.Sequential(
    # Layer. de Output
    [tf.keras.layers.Dense(1, input_shape=(1,))]
)
```

```python
model_4_lista = tf.keras.models.Sequential()
```

```python
model_4_lista.add( tf.keras.layers.Dense(1, input_shape=(1,)))
```

```python
model_4_secvential.summary()
```

```python
model_4_lista.summary()
```

```python
# Varianta desclarativa
model_4_secvential.compile(loss='mse', optimizer='sgd', metrics=['mae'])
```

```python
# Varianta explicita
model_4_lista.compile(loss=tf.keras.losses.mse, optimizer=tf.keras.optimizers.SGD(learning_rate=0.01), metrics=['mae'])
```

Imbunatatirea modelului - exista anumite elemente pe care le pot schimba pentru a face modulul mai performant

# 1. Adaugare mai multor straturi ( layere)

# 2. Adaugarea mai multor neurori ( hidden units)

# 3. Schimbarea functiei de activare

# 4. Schimbare optimizatorului

# 5. Schimbarea learning rate-utului din optimizator

# 6. Adaugarea mai multor date

# 7. Antrenarea pe o perioada mai lunga
