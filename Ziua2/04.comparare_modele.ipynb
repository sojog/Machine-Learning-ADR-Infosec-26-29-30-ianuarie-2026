```python
import numpy as np                  # calcul numeric
import matplotlib.pyplot as plt     # vizualizare

from sklearn.datasets import make_regression

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

from sklearn.model_selection import train_test_split
```

```python
x, y = make_regression(n_samples=100, n_features=1, noise=10)
```

```python
plt.scatter(x, y)
plt.xlabel("Suprafete")
plt.ylabel("Preturi")
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
plt.scatter(x_train, y_train)
plt.scatter(x_test, y_test)
plt.xlabel("Suprafete")
plt.ylabel("Preturi")
```

# Ca sa pot compara -> impart datele de antrenament in test si train

```python
linear_model = LinearRegression()
linear_model
```

```python
knn_model = KNeighborsRegressor()
knn_model
```

```python
linear_model.fit(x_train, y_train)
```

```python
knn_model.fit(x_train, y_train)
```

```python
linear_model.score(x_test, y_test)
```

```python
knn_model.score(x_test, y_test)
```

```python
linear_model.predict([[55]])
```

```python
knn_model.predict([[55]])
```

```python
y_prezis_linear = linear_model.predict(x_test)
y_prezis_linear
```

```python
y_prezis_knn = knn_model.predict(x_test)
y_prezis_knn
```
