```python
import numpy as np                  # calcul numeric
import matplotlib.pyplot as plt     # vizualizare

from sklearn.datasets import make_regression

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
apartamente = np.array([(40, 60_000), (50, 75_000), (60, 97_000), (65, 88_000), (70, 105_000), (80, 125_000),(120, 300_000)])
```

```python
suprafete = apartamente[ : , 0  ]
preturi = apartamente [ : , 1]

plt.scatter(suprafete, preturi)
plt.xlabel("Suprafete")
plt.ylabel("Preturi")
```

```python
linear_model = LinearRegression()
linear_model
```

```python
knn_model = KNeighborsRegressor()
knn_model
```

```python
suprafete
```

```python
suprafete_2d = suprafete.reshape(-1, 1)
suprafete_2d
```

```python
linear_model.fit(suprafete_2d, preturi)
```

```python
knn_model.fit(suprafete_2d, preturi)
```

```python
plt.scatter(suprafete, preturi)
plt.xlabel("Suprafete")
plt.ylabel("Preturi")
```

```python
linear_model.predict([[55]])
```
