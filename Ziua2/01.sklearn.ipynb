# KNN. - K Neirest Neighbors

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
x, y = make_regression(n_samples=1000, n_features=1, noise=10)

model = LinearRegression()
model.fit(x, y)

plt.scatter(x, y)
plt.plot(x, model.coef_ * x + model.intercept_)
```

```python
x, y = make_regression(n_samples=10, n_features=1, noise=1)

model = LinearRegression()
model.fit(x, y)

plt.scatter(x, y)
plt.plot(x, model.coef_ * x + model.intercept_)
```
