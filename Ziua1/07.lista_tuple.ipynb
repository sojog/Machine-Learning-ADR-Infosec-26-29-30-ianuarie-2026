# Lista - mutabila (se pot adauga, sterge, edita)
# Tuplu - imutabil (nu se pot adauga, sterge, edita)

```python
o_lista = [1, 2, 3, "salut", True, 3.0]
un_tuplu = (1, 2, 3, "salut", True, 3.0)
```

```python
o_lista.append(20)
o_lista
```

```python
o_lista.un_tuplu(20)
```

```python
o_lista.pop()
o_lista
```

```python
o_lista.pop(2)
o_lista
```

```python
un_tuplu.pop(2)
```

```python
o_lista.clear()
```

```python
o_lista
```

! Tuplu poate fi redefinit (dar nu inseamna ca este mutabil)

```python
un_tuplu = (1, 2, 3, "salut", True, 3.0)
un_tuplu = (1, 2, 3, "salut", True)
```
