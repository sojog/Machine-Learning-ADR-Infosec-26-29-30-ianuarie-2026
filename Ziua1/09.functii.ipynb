# Parametri optionali

```python
def suma(*args):
    print(args, type(args))
```

```python
suma()
```

```python
suma(2, 3, 4)
```

```python
suma(2, 3, 4, 4132, 312, 31, 2, 312)
```

```python
def suma(*args, **kargs):
    print(args, type(args))
    print(kargs, type(kargs))
```

```python
suma(2, 3, 4, 4132, 312, 31, 2, 312)
```

```python
suma(2, 3, 4, 4132, 312, 31, 2, 312, a=32, b=54)
```
