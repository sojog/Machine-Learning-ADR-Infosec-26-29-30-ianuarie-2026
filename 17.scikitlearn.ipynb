```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
apartamente = np.array([(40, 60_000), (50, 75_000), (60, 97_000), (65, 88_000), (70, 105_000), (80, 125_000),(120, 300_000)])
suprafete = apartamente[ : , 0  ]
preturile = apartamente[ : , 1  ]
```

```python
plt.scatter(suprafete, preturile);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
suprafete 
```

```python
suprafete.shape
```

# [ 40,   50,  60,  65,  70,  80, 120] -> [[ 40],  [50],  [60],  [65],  [70],  [80], [120]]]

```python
suprafete_2D = suprafete.reshape(-1, 1)
suprafete_2D
```

```python
model.fit(suprafete_2D, preturile)
```

```python
model.predict([ [100], [130]  ])
```
