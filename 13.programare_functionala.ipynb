# 1. Map

```python
lista = [2, 3, 4, 5]
```

```python
def adunare(x):
    return x + 2
```

```python
map( adunare, lista)
```

```python
list(map( adunare, lista))
```

```python
list(map( lambda x: x + 2, lista))
```

# 2. Filter

```python
lista = [2, 3, 4, 5]
filter(lambda x: x > 3, lista)
```

```python
list(filter(lambda x: x > 3, lista))
```

# 3. Reduce

```python
from functools import reduce
```

```python
lista = [2, 3, 4, 5]
reduce(lambda x, y: x + y , lista)
```
