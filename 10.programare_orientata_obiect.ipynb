# Clasa este un sablon

```python
# Conventie - > numele clasei cu litera mare
class Elev:
    # variabila in clasa se numeste atribut
    varsta = 10
    # functia in clasa se numeste metoda
    def este_major(varsta):
        if varsta > 18:
            return True
        else:
            return False
```

# Obiectul este o instanta (concretizare a clasei)

```python
obiect1 = Elev()
```

```python
# Conventie - > numele clasei cu litera mare
class Elev:

    # init -> initializare
    def __init__(self, varsta):
        # variabila in clasa se numeste atribut
        self.varsta = varsta


    # functia in clasa se numeste metoda
    def este_major(self):
        if self.varsta > 18:
            return True
        else:
            return False
```

```python
obiect1 = Elev(17)
obiect1
```

```python
obiect2 = Elev(27)
obiect2
```

```python
obiect1.este_major()
```

```python
obiect2.este_major()
```
